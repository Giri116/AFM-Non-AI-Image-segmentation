```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# try sklearn, fallback to cv2.kmeans if sklearn not available
try:
    from sklearn.cluster import KMeans
    _HAVE_SKLEARN = True
except Exception:
    _HAVE_SKLEARN = False
    import cv2

def kmeans_binary_masks(img_input, k=3, show=True, random_state=42):
    """
    Perform KMeans (color) clustering on an RGB image and produce binary masks for each cluster.

    Parameters
    ----------
    img_input : str or numpy.ndarray
        Path to an image file (e.g. "Image1.jpg") or an HxWx3 uint8 RGB numpy array.
    k : int
        Number of clusters (default 3).
    show : bool
        If True, display original image and binary masks with matplotlib.
    random_state : int
        Random state for reproducibility (when using sklearn).

    Returns
    -------
    results : dict
        {
            'image': original_image (H,W,3 uint8),
            'labels': labels_img (H,W int, values in 0..k-1),
            'masks': masks (k, H, W) uint8 with values {0,1},
            'centers': centers (k,3) uint8 - cluster center colors
        }
    """
    # load image
    if isinstance(img_input, str):
        img = Image.open(img_input).convert("RGB")
        img_np = np.array(img)
    elif isinstance(img_input, np.ndarray):
        img_np = img_input.copy()
        if img_np.dtype != np.uint8:
            img_np = (np.clip(img_np, 0, 1) * 255).astype(np.uint8) if img_np.max() <= 1.1 else img_np.astype(np.uint8)
        if img_np.ndim != 3 or img_np.shape[2] != 3:
            raise ValueError("If passing a numpy array, it must be HxWx3 (RGB).")
    else:
        raise TypeError("img_input must be a file path or an HxWx3 numpy array.")

    h, w, _ = img_np.shape
    pixels = img_np.reshape(-1, 3).astype(np.float64)

    # run k-means
    if _HAVE_SKLEARN:
        kmeans = KMeans(n_clusters=k, random_state=random_state, n_init=10)
        labels_flat = kmeans.fit_predict(pixels)
        centers = kmeans.cluster_centers_.round().astype(np.uint8)
    else:
        Z = pixels.astype(np.float32)
        criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)
        attempts = 10
        _, labels_flat, centers = cv2.kmeans(Z, k, None, criteria, attempts, cv2.KMEANS_RANDOM_CENTERS)
        labels_flat = labels_flat.flatten().astype(int)
        centers = centers.astype(np.uint8)

    labels_img = labels_flat.reshape(h, w)

    # create binary masks (0/1 uint8)
    masks = np.stack([(labels_img == i).astype(np.uint8) for i in range(k)], axis=0)  # (k, h, w)

    # show with matplotlib
    if show:
        import math
        total = 1 + k
        cols = min(4, total)
        rows = math.ceil(total / cols)
        fig = plt.figure(figsize=(4 * cols, 4 * rows))
        ax = fig.add_subplot(rows, cols, 1)
        ax.imshow(img_np)
        ax.set_title("Original")
        ax.axis("off")

        for i in range(k):
            ax = fig.add_subplot(rows, cols, 2 + i)
            # show mask as black/white by mapping 0->0 and 1->255 in RGB to avoid colormap
            mask_rgb = np.stack([masks[i] * 255]*3, axis=2).astype(np.uint8)
            ax.imshow(mask_rgb)
            ax.set_title(f"Mask cluster {i} (1=fg,0=bg)")
            ax.axis("off")

        plt.tight_layout()
        plt.show()

    return {
        'image': img_np,
        'labels': labels_img,
        'masks': masks,
        'centers': centers
    }
```

```python
import os
import io
import numpy as np
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt

# Try sklearn; fallback to cv2.kmeans
try:
    from sklearn.cluster import KMeans
    _HAVE_SKLEARN = True
except Exception:
    _HAVE_SKLEARN = False
    import cv2
    _HAVE_SKLEARN = False

# Helper to fetch Roboto if not present
def _ensure_roboto(dest_path="Roboto-Regular.ttf"):
    """
    Ensure Roboto-Regular.ttf exists at dest_path. If not, attempt to download
    from the Google Fonts GitHub repository. Returns the path to the TTF file.
    If download fails, returns None.
    """
    if os.path.exists(dest_path):
        return dest_path
    try:
        import urllib.request
        url = "https://github.com/google/fonts/raw/main/apache/roboto/Roboto-Regular.ttf"
        print(f"Downloading Roboto font from {url} ...")
        urllib.request.urlretrieve(url, dest_path)
        print("Downloaded Roboto to", dest_path)
        return dest_path
    except Exception as e:
        print("Could not download Roboto font automatically:", str(e))
        return None

def kmeans_area_percentages_with_annotation(img_input, k=3, show=True, font_path=None, font_size_ratio=0.035, random_state=42):
    """
    Compute area percentages for k color clusters (k must be 2 or 3), annotate image with Roboto font.

    Parameters
    ----------
    img_input : str or numpy.ndarray
        Path to an image file (e.g. "Image1.jpg") or an HxWx3 uint8 RGB numpy array.
    k : int
        Number of clusters to compute. Should be 2 or 3 per your request.
    show : bool
        If True, display the annotated image using matplotlib.
    font_path : str or None
        Optional path to a .ttf Roboto font. If None the function will try to download Roboto-Regular.ttf
        into the current working directory. If font cannot be loaded, uses default PIL font.
    font_size_ratio : float
        Font size relative to image height (approx). Smaller values for large images.
    random_state : int
        Random seed for KMeans (when using sklearn).

    Returns
    -------
    result : dict
        {
          'percentages': list of floats (percentage for each cluster, length k),
          'labels': HxW int array of cluster indices,
          'masks': (k, H, W) uint8 arrays with values {0,1},
          'annotated_image': PIL.Image annotated copy (RGB),
          'centers': (k,3) uint8 cluster center colors
        }
    """
    if k not in (2, 3):
        raise ValueError("k must be 2 or 3 as requested.")

    # load image
    if isinstance(img_input, str):
        if not os.path.exists(img_input):
            raise FileNotFoundError(f"Input image not found: {img_input}")
        img = Image.open(img_input).convert("RGB")
        img_np = np.array(img)
    elif isinstance(img_input, np.ndarray):
        img_np = img_input.copy()
        if img_np.dtype != np.uint8:
            img_np = (np.clip(img_np, 0, 1) * 255).astype(np.uint8) if img_np.max() <= 1.1 else img_np.astype(np.uint8)
        if img_np.ndim != 3 or img_np.shape[2] != 3:
            raise ValueError("If passing a numpy array, it must be HxWx3 (RGB).")
        img = Image.fromarray(img_np)
    else:
        raise TypeError("img_input must be a file path or an HxWx3 numpy array.")

    h, w, _ = img_np.shape
    pixels = img_np.reshape(-1, 3).astype(np.float64)

    # run KMeans on pixel colors
    if _HAVE_SKLEARN:
        kmeans = KMeans(n_clusters=k, random_state=random_state, n_init=10)
        labels_flat = kmeans.fit_predict(pixels)
        centers = kmeans.cluster_centers_.round().astype(np.uint8)
    else:
        # cv2.kmeans fallback
        import cv2
        Z = pixels.astype(np.float32)
        criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)
        attempts = 10
        _, labels_flat, centers = cv2.kmeans(Z, k, None, criteria, attempts, cv2.KMEANS_RANDOM_CENTERS)
        labels_flat = labels_flat.flatten().astype(int)
        centers = centers.astype(np.uint8)

    labels = labels_flat.reshape(h, w)

    # compute masks and percentages
    masks = np.stack([(labels == i).astype(np.uint8) for i in range(k)], axis=0)
    total_pixels = h * w
    percentages = [float(masks[i].sum()) / total_pixels * 100.0 for i in range(k)]

    # prepare annotation text lines
    lines = []
    for i, pct in enumerate(percentages):
        # round to 2 decimals, show cluster index and percentage
        lines.append(f"Cluster {i}: {pct:.2f}%")

    # Load Roboto font (try provided path, else try to download)
    font = None
    if font_path:
        try:
            font = ImageFont.truetype(font_path, max(12, int(h * font_size_ratio)))
        except Exception as e:
            print("Could not load provided font_path:", e)
            font = None

    if font is None:
        # try common local filenames
        possible = ["Roboto-Regular.ttf", os.path.expanduser("~/Roboto-Regular.ttf")]
        found = None
        for p in possible:
            if p and os.path.exists(p):
                found = p
                break
        if found is None:
            # attempt automatic download
            found = _ensure_roboto("Roboto-Regular.ttf")
        if found:
            try:
                font = ImageFont.truetype(found, max(12, int(h * font_size_ratio)))
            except Exception as e:
                print("Failed to load downloaded/system Roboto font:", e)
                font = None

    if font is None:
        # fallback to default PIL font
        font = ImageFont.load_default()
        print("Using default font (Roboto unavailable).")

    # annotate a copy of original image
    annotated = img.copy()
    draw = ImageDraw.Draw(annotated)

    # Draw a semi-opaque rectangle as a background for readability
    # compute rectangle size depending on text lines
    padding = int(max(8, h * 0.01))
    line_heights = [draw.textsize(line, font=font)[1] for line in lines]
    text_height = sum(line_heights) + padding * (len(lines)+1)
    text_width = max(draw.textsize(line, font=font)[0] for line in lines) + padding*2

    # place rectangle in top-left corner with margin
    margin = int(max(6, h * 0.02))
    rect_x0, rect_y0 = margin, margin
    rect_x1, rect_y1 = rect_x0 + text_width, rect_y0 + text_height

    # semi-transparent black rectangle (PIL doesn't support alpha on RGB image directly: use overlay)
    overlay = Image.new("RGBA", annotated.size, (255,255,255,0))
    ov_draw = ImageDraw.Draw(overlay)
    ov_draw.rectangle([rect_x0, rect_y0, rect_x1, rect_y1], fill=(0,0,0,150))  # semi-transparent black
    annotated = Image.alpha_composite(annotated.convert("RGBA"), overlay).convert("RGB")
    draw = ImageDraw.Draw(annotated)

    # write lines in white with small left margin
    y = rect_y0 + padding
    for line in lines:
        draw.text((rect_x0 + padding, y), line, font=font, fill=(255,255,255))
        y += draw.textsize(line, font=font)[1] + int(padding/2)

    # also optionally draw small colored swatches for each cluster next to text
    swatch_size = int(max(12, h * 0.02))
    # redraw overlay to place swatches (we have annotated; use draw)
    y = rect_y0 + padding
    for i in range(k):
        sw_x0 = rect_x0 + 2
        sw_y0 = y
        sw_x1 = sw_x0 + swatch_size
        sw_y1 = sw_y0 + swatch_size
        draw.rectangle([sw_x0, sw_y0, sw_x1, sw_y1], fill=tuple(int(c) for c in centers[i].tolist()))
        # move y by approx line height
        y += draw.textsize(lines[i], font=font)[1] + int(padding/2)

    # display annotated image and optionally masks
    if show:
        import math
        total = 1 + k
        cols = min(4, total)
        rows = math.ceil(total / cols)
        fig = plt.figure(figsize=(4 * cols, 4 * rows))
        ax = fig.add_subplot(rows, cols, 1)
        ax.imshow(np.asarray(annotated))
        ax.set_title("Annotated image (percentages)")
        ax.axis("off")

        for i in range(k):
            ax = fig.add_subplot(rows, cols, 2 + i)
            mask_rgb = np.stack([masks[i] * 255]*3, axis=2).astype(np.uint8)
            ax.imshow(mask_rgb)
            ax.set_title(f"Cluster {i} mask\n{percentages[i]:.2f}%")
            ax.axis("off")
        plt.tight_layout()
        plt.show()

    result = {
        "percentages": percentages,
        "labels": labels,
        "masks": masks,
        "annotated_image": annotated,
        "centers": centers
    }
    return result

# Example usage (uncomment in Jupyter):
# res = kmeans_area_percentages_with_annotation("Image1.jpg", k=3, show=True)
# print("Percentages:", res['percentages'])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from sklearn.cluster import KMeans

def intensity_area_percentage(image_path, k=3, show=True):
    """
    Calculates area percentage of regions based on 2 or 3 intensity clusters.

    Parameters:
    -----------
    image_path : str
        Path to input image (any color image or grayscale).
    k : int
        Number of clusters (2 or 3).
    show : bool
        If True, display clustered image with percentages overlayed.

    Returns:
    --------
    percentages : list
        Area percentages of each intensity cluster.
    """

    # load image and convert to grayscale
    img = Image.open(image_path).convert('L')  # grayscale
    img_np = np.array(img)
    h, w = img_np.shape

    # reshape for KMeans
    pixels = img_np.reshape(-1, 1).astype(np.float64)

    # run KMeans on intensity values
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(pixels)
    labels_img = labels.reshape(h, w)

    # compute area percentages
    total_pixels = h * w
    percentages = [(labels == i).sum() / total_pixels * 100 for i in range(k)]

    # print results
    print("Intensity Cluster Area Percentages:")
    for i, pct in enumerate(percentages):
        print(f"Cluster {i}: {pct:.2f}%")

    if show:
        # create a color visualization of clusters
        cluster_colors = np.linspace(0, 255, k).astype(np.uint8)
        clustered_img = np.zeros_like(img_np)
        for i in range(k):
            clustered_img[labels_img == i] = cluster_colors[i]

        plt.figure(figsize=(6, 5))
        plt.imshow(clustered_img, cmap='gray')
        plt.axis('off')
        plt.title("Clustered Image (by intensity)")

        # show percentages on the image
        text_y = 20
        for i, pct in enumerate(percentages):
            plt.text(10, text_y, f"Cluster {i}: {pct:.2f}%", color='red', fontsize=6, weight='bold')
            text_y += 20

        plt.show()

    return percentages
```
