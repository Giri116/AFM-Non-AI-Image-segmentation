```python
from PIL import Image as Im,ImageOps
import numpy as np
from matplotlib import pyplot as plt
im=Im.open("4d.jpg")
arr=np.array(im)
arr=ImageOps.grayscale(im)
arr=np.array(arr)
grayarr=arr
from collections import defaultdict
D=defaultdict(int)
for i in range(256):
    D[i]=0
l,b=np.shape(arr)
for i in range(l):
    for j in range(b):
        D[arr[i][j]]+=1
L_x=[]
L_y=[]
for x in D:
    L_x.append(x)
    L_y.append(D[x])
fig = plt.figure(figsize =(60, 17))#figsize =(10, 10)
plt.bar(L_x,L_y)
plt.show()
```

```python
import cv2
p,q = grayarr.shape
fg = np.zeros((p,q,3),dtype=np.uint8)
for i in range(p):
    for j in range(q):
        inte = grayarr[i][j]
        fg[i,j] = [inte,inte,inte]
im_v = cv2.hconcat([np.array(im), np.array(fg)])
Im.fromarray(im_v).save("oandg.jpg")
```
