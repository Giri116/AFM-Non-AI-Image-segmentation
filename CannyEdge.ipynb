```python
import cv2
import matplotlib.pyplot as plt

# Read the image
image = cv2.imread('Image1.jpg')

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply bilateral filter for edge-preserving smoothing
# Parameters: (image, diameter, sigmaColor, sigmaSpace)
blurred = cv2.bilateralFilter(gray, d=9, sigmaColor=75, sigmaSpace=75)

# Apply Canny edge detection
edges = cv2.Canny(blurred, 100, 200)

# Display the original, blurred, and Canny edge images
plt.figure(figsize=(15, 5))

# Original image (convert BGR to RGB for proper color display)
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Bilateral filtered image
plt.subplot(1, 3, 2)
plt.imshow(blurred, cmap='gray')
plt.title('Bilateral Filtered Image')
plt.axis('off')

# Canny edges
plt.subplot(1, 3, 3)
plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Read the image
image = cv2.imread('Image1.jpg')

# Convert BGR (OpenCV default) to RGB for correct processing
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Apply bilateral filter to each RGB channel
filtered_rgb = cv2.bilateralFilter(image_rgb, d=11, sigmaColor=75, sigmaSpace=75)

# Convert the filtered image to grayscale for Canny edge detection
gray_filtered = cv2.cvtColor(filtered_rgb, cv2.COLOR_RGB2GRAY)

# Apply Canny edge detection
edges = cv2.Canny(gray_filtered, 100, 200)

# Display results
plt.figure(figsize=(15, 5))

# Original image
plt.subplot(1, 3, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

# RGB Bilateral Filtered image
plt.subplot(1, 3, 2)
plt.imshow(filtered_rgb)
plt.title('RGB Bilateral Filtered Image')
plt.axis('off')

# Canny edges
plt.subplot(1, 3, 3)
plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')

plt.tight_layout()
plt.show()
```
