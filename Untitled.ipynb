```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def rgb_unit_normalization(image_path):
    # Read image in RGB
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = img.astype(np.float64)

    # Compute magnitude of RGB vector for each pixel
    magnitude = np.sqrt(np.sum(img ** 2, axis=2, keepdims=True)) + 1e-12  # Avoid divide by zero

    # Normalize each RGB vector to unit magnitude
    normalized_img = (img / magnitude)

    # Compute grayscale image from magnitude
    gray_magnitude = magnitude.squeeze() / np.max(magnitude)  # Normalize to 01 for display

    # Display images
    plt.figure(figsize=(12, 4))
    
    plt.subplot(1, 3, 1)
    plt.imshow(np.clip(img / 255.0, 0, 1))
    plt.title("Original RGB Image")
    plt.axis('off')
    
    plt.subplot(1, 3, 2)
    plt.imshow(np.clip(normalized_img, 0, 1))
    plt.title("Unit Magnitude Normalized RGB")
    plt.axis('off')
    
    plt.subplot(1, 3, 3)
    plt.imshow(gray_magnitude, cmap='gray')
    plt.title("Grayscale Magnitude Image")
    plt.axis('off')
    
    plt.tight_layout()
    plt.show()
    return normalized_img,gray_magnitude

# Example usage:
# rgb_unit_normalization('lena.png')
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def rgb_to_hsi(image_path):
    # Read the image in RGB
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = img.astype(np.float64) / 255.0  # Normalize to [0,1]
    
    R, G, B = img[:,:,0], img[:,:,1], img[:,:,2]
    
    # Intensity component
    I = (R + G + B) / 3.0
    
    # Saturation component
    min_rgb = np.minimum(np.minimum(R, G), B)
    S = 1 - (3 / (R + G + B + 1e-12)) * min_rgb
    
    # Hue component
    numerator = 0.5 * ((R - G) + (R - B))
    denominator = np.sqrt((R - G)**2 + (R - B)*(G - B)) + 1e-12
    theta = np.arccos(np.clip(numerator / denominator, -1, 1))
    
    H = np.zeros_like(R)
    H[B <= G] = theta[B <= G]
    H[B > G] = 2 * np.pi - theta[B > G]
    H = H / (2 * np.pi)  # Normalize hue to [0,1]
    
    # Display results
    plt.figure(figsize=(12, 4))
    
    plt.subplot(1, 4, 1)
    plt.imshow(img)
    plt.title('Original RGB')
    plt.axis('off')
    
    plt.subplot(1, 4, 2)
    plt.imshow(H, cmap='gray')
    plt.title('Hue')
    plt.axis('off')
    
    plt.subplot(1, 4, 3)
    plt.imshow(S, cmap='gray')
    plt.title('Saturation')
    plt.axis('off')
    
    plt.subplot(1, 4, 4)
    plt.imshow(I, cmap='gray')
    plt.title('Intensity')
    plt.axis('off')
    
    plt.tight_layout()
    plt.show()
    
    return H, S, I

# Example usage:
# H, S, I = rgb_to_hsi('lena.png')
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def pseudo_color_image(image_path, w=5):
    # Read grayscale image
    img_gray = image_path #cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    img_gray = img_gray.astype(np.float64)

    # Normalize intensities to [0, 255]
    img_gray = (img_gray - img_gray.min()) / (img_gray.max() - img_gray.min() + 1e-12) * 255
    H, W = img_gray.shape

    # Define window boundaries
    bins = np.linspace(0, 256, w+1)

    # Define distinct colors for each window (R,G,B)
    colors = np.array([
        [255, 0, 0],     # Red
        [0, 255, 0],     # Green
        [0, 0, 255],     # Blue
        [255, 255, 0],   # Yellow
        [255, 0, 255],   # Magenta
        [0, 255, 255],   # Cyan
        [255, 128, 0],   # Orange
        [128, 0, 255],   # Violet
    ], dtype=np.float64)

    # Clip colors if w > len(colors)
    if w > len(colors):
        # Generate random colors if needed
        np.random.seed(42)
        colors = np.random.randint(0, 255, (w, 3)).astype(np.float64)

    # Initialize output pseudo-color image
    pseudo_img = np.zeros((H, W, 3), dtype=np.float64)

    # Assign colors based on intensity window
    for i in range(w):
        mask = (img_gray >= bins[i]) & (img_gray < bins[i+1])
        pseudo_img[mask] = colors[i % len(colors)]

    # Normalize to [0,1] for display
    pseudo_img = pseudo_img / 255.0

    # Display results
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    plt.imshow(img_gray, cmap='gray')
    plt.title('Original Grayscale Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(pseudo_img)
    plt.title(f'Pseudo-colored Image (w={w})')
    plt.axis('off')

    plt.tight_layout()
    plt.show()

    return pseudo_img

# Example usage:
# pseudo_img = pseudo_color_image('lena_gray.png', w=6)
```

```python
#!/usr/bin/env python3
"""
STM segmentation comparison (non-AI)
Methods: OpenCV KMeans, Marker-based Watershed, SLIC+GMM (superpixel merging)
Also splits 'product' region into cis/trans using gradient orientation clustering.
Saves overlays and a CSV summary with areas.
"""

import os, argparse
import numpy as np
import cv2
from skimage import filters, morphology, segmentation, color, exposure
from skimage.segmentation import slic
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
import scipy.ndimage as ndi
import pandas as pd

def norm01(x):
    x = x - x.min()
    if x.max()>0:
        x = x / x.max()
    return x

def preprocess(img_gray, bg_sigma=50, denoise_h=10):
    # background removal
    bg = cv2.GaussianBlur(img_gray, (0,0), sigmaX=bg_sigma, sigmaY=bg_sigma)
    img_corr = img_gray - bg
    img_corr = img_corr - img_corr.min()
    if img_corr.max() > 0:
        img_corr = img_corr / img_corr.max()
    # denoise
    img_dn = cv2.fastNlMeansDenoising((img_corr*255).astype(np.uint8), None, h=denoise_h, templateWindowSize=7, searchWindowSize=21).astype(np.float32)/255.0
    # slight unsharp to enhance faint edges
    gauss = cv2.GaussianBlur(img_dn, (0,0), sigmaX=1)
    img_sharp = cv2.addWeighted(img_dn, 1.5, gauss, -0.5, 0.0)
    return img_sharp

def compute_features(img_sharp):
    sobx = cv2.Sobel(img_sharp, cv2.CV_32F, 1, 0, ksize=3)
    soby = cv2.Sobel(img_sharp, cv2.CV_32F, 0, 1, ksize=3)
    gradmag = np.hypot(sobx, soby)
    mean = cv2.GaussianBlur(img_sharp, (7,7), 0)
    mean_sq = cv2.GaussianBlur(img_sharp*img_sharp, (7,7), 0)
    local_std = np.sqrt(np.clip(mean_sq - mean*mean, 0, None))
    return norm01(img_sharp), norm01(gradmag), norm01(local_std), sobx, soby

def orientation_split(product_mask, sobx, soby):
    # split product into 2 orientation clusters (cis/trans) using angle on unit circle
    angles = np.arctan2(soby, sobx)
    coords = np.argwhere(product_mask)
    if coords.shape[0] == 0:
        return np.zeros_like(product_mask), np.zeros_like(product_mask)
    ang_vals = angles[product_mask].reshape(-1,1)
    X = np.concatenate([np.cos(ang_vals), np.sin(ang_vals)], axis=1).astype(np.float32)
    # small KMeans on the orientation vectors
    k = KMeans(n_clusters=2, random_state=0).fit(X)
    lbls = k.labels_
    cis = np.zeros_like(product_mask, dtype=bool)
    trans = np.zeros_like(product_mask, dtype=bool)
    cis[coords[lbls==0][:,0], coords[lbls==0][:,1]] = True
    trans[coords[lbls==1][:,0], coords[lbls==1][:,1]] = True
    return cis, trans

def overlay_and_save(base_rgb, cis_mask, trans_mask, holes_mask, outpath):
    over = base_rgb.copy().astype(np.float32)
    mask_rgb = base_rgb.copy()
    mask_rgb[cis_mask] = (0,255,255)      # cyan for cis
    mask_rgb[trans_mask] = (255,0,255)    # magenta for trans
    mask_rgb[holes_mask] = (255,0,0)      # red for holes
    alpha = 0.5
    vis = (over*(1-alpha) + mask_rgb*alpha).astype(np.uint8)
    cv2.imwrite(outpath, cv2.cvtColor(vis, cv2.COLOR_RGB2BGR))
    return outpath

def method_kmeans(img_sharp, f_grad, f_std):
    H,W = img_sharp.shape
    X = np.stack([img_sharp.ravel(), f_grad.ravel(), f_std.ravel()], axis=1).astype(np.float32)
    # use OpenCV kmeans for speed
    criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)
    ret, labels, centers = cv2.kmeans(X, 3, None, criteria, 8, cv2.KMEANS_PP_CENTERS)
    lbl_img = labels.flatten().reshape(H,W)
    # map clusters by intensity ascending: 0 darkest->holes, 1 mid->product, 2 brightest->background
    means = [img_sharp[lbl_img==k].mean() if (lbl_img==k).any() else 0 for k in range(3)]
    order = np.argsort(means)
    mapped = np.zeros_like(lbl_img)
    for i,k in enumerate(order):
        mapped[lbl_img==k] = i
    holes = (mapped==0)
    product = (mapped==1)
    background = (mapped==2)
    product = morphology.remove_small_objects(product, min_size=80)
    product = morphology.binary_closing(product, morphology.disk(3))
    product = ndi.binary_fill_holes(product)
    return holes, product, background

def method_watershed(img_sharp, f_grad):
    # choose percentile based markers
    p10 = np.percentile(img_sharp, 10)
    p90 = np.percentile(img_sharp, 90)
    markers = np.zeros_like(img_sharp, dtype=np.int32)
    markers[img_sharp < p10] = 1
    markers[img_sharp > p90] = 3
    markers[(img_sharp>=p10) & (img_sharp<=p90)] = 2
    grad = filters.sobel(img_sharp)
    ws = segmentation.watershed(grad, markers=markers, compactness=0.001)
    holes = (ws==1)
    product = (ws==2)
    background = (ws==3)
    product = morphology.remove_small_objects(product, min_size=80)
    product = morphology.binary_closing(product, morphology.disk(3))
    product = ndi.binary_fill_holes(product)
    return holes, product, background

def method_slic_gmm(rgb, img_sharp, f_grad, f_std, n_segments=700):
    segments = slic(rgb, n_segments=n_segments, compactness=10, sigma=1, start_label=1)
    n_seg = segments.max()
    Xsp = np.zeros((n_seg, 3), dtype=np.float32)
    for s in range(1, n_seg+1):
        mask = (segments==s)
        if mask.sum()==0:
            continue
        Xsp[s-1,0] = img_sharp[mask].mean()
        Xsp[s-1,1] = f_grad[mask].mean()
        Xsp[s-1,2] = f_std[mask].mean()
    gmm = GaussianMixture(n_components=3, random_state=0).fit(Xsp)
    sp_lbl = gmm.predict(Xsp)
    lbl_map = np.zeros(img_sharp.shape, dtype=np.int32)
    for s in range(1, n_seg+1):
        lbl_map[segments==s] = int(sp_lbl[s-1])
    # map by mean intensity ascending
    means = [img_sharp[lbl_map==k].mean() if (lbl_map==k).any() else 0 for k in range(3)]
    order = np.argsort(means)
    mapped = np.zeros_like(lbl_map)
    for i,k in enumerate(order):
        mapped[lbl_map==k] = i
    holes = (mapped==0)
    product = (mapped==1)
    background = (mapped==2)
    product = morphology.remove_small_objects(product, min_size=80)
    product = morphology.binary_closing(product, morphology.disk(3))
    product = ndi.binary_fill_holes(product)
    return holes, product, background

def main(args):
    img = cv2.imread(args.input, cv2.IMREAD_COLOR)
    if img is None:
        raise RuntimeError("Could not read input image")
    rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY).astype(np.float32)/255.0

    img_sharp = preprocess(gray, bg_sigma=args.bg_sigma, denoise_h=args.denoise_h)
    f_int, f_grad, f_std, sobx, soby = compute_features(img_sharp)

    os.makedirs(args.outdir, exist_ok=True)

    table = []
    # KMeans pixel
    holes, product, background = method_kmeans(f_int, f_grad, f_std)
    cis, trans = orientation_split(product, sobx, soby)
    overlay_path = os.path.join(args.outdir, "overlay_kmeans.png")
    overlay_and_save(rgb, cis, trans, holes, overlay_path)
    total = product.sum() + holes.sum() + background.sum()
    table.append({"method":"KMeans", "cis_px": int(cis.sum()), "trans_px": int(trans.sum()), "holes_px": int(holes.sum()), "total_px": int(total)})

    # Watershed
    holes, product, background = method_watershed(f_int, f_grad)
    cis, trans = orientation_split(product, sobx, soby)
    overlay_path = os.path.join(args.outdir, "overlay_watershed.png")
    overlay_and_save(rgb, cis, trans, holes, overlay_path)
    total = product.sum() + holes.sum() + background.sum()
    table.append({"method":"Watershed", "cis_px": int(cis.sum()), "trans_px": int(trans.sum()), "holes_px": int(holes.sum()), "total_px": int(total)})

    # SLIC + GMM
    holes, product, background = method_slic_gmm(rgb, f_int, f_grad, f_std, n_segments=args.slic_segments)
    cis, trans = orientation_split(product, sobx, soby)
    overlay_path = os.path.join(args.outdir, "overlay_slic_gmm.png")
    overlay_and_save(rgb, cis, trans, holes, overlay_path)
    total = product.sum() + holes.sum() + background.sum()
    table.append({"method":"SLIC+GMM", "cis_px": int(cis.sum()), "trans_px": int(trans.sum()), "holes_px": int(holes.sum()), "total_px": int(total)})

    df = pd.DataFrame(table)
    # add percents
    df["cis_%"] = (df["cis_px"] / df["total_px"] * 100).round(3)
    df["trans_%"] = (df["trans_px"] / df["total_px"] * 100).round(3)
    df["holes_%"] = (df["holes_px"] / df["total_px"] * 100).round(3)
    csv_path = os.path.join(args.outdir, "segmentation_summary.csv")
    df.to_csv(csv_path, index=False)
    print("Saved overlays and summary to", args.outdir)
    print(df)
    return

if __name__ == "__main__":
    p = argparse.ArgumentParser()
    p.add_argument("--input",default="Image1.jpg", required=True, help="input image path")
    p.add_argument("--outdir", default="outputs", help="output directory")
    p.add_argument("--bg-sigma", type=int, default=50, dest="bg_sigma", help="background Gaussian sigma")
    p.add_argument("--denoise-h", type=float, default=10.0, dest="denoise_h", help="NLM h parameter")
    p.add_argument("--slic-segments", type=int, default=700, dest="slic_segments", help="SLIC n_segments")
    args = p.parse_args()
    main(args)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def masked_otsu_threshold(gray_path, mask_path):
    # Read grayscale image and mask
    gray = gray_path#cv2.imread(gray_path, cv2.IMREAD_GRAYSCALE)
    mask = mask_path#cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)

    # Extract only masked pixels
    masked_pixels = gray[mask == 255]

    # Compute Otsu threshold on masked region
    t, _ = cv2.threshold(masked_pixels, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Apply threshold only on masked region
    thresh = np.zeros_like(gray)
    thresh[(gray >= t) & (mask == 255)] = 255

    # Display results
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 3, 1), plt.imshow(gray, cmap='gray'), plt.title('Original Gray'), plt.axis('off')
    plt.subplot(1, 3, 2), plt.imshow(mask, cmap='gray'), plt.title('Binary Mask'), plt.axis('off')
    plt.subplot(1, 3, 3), plt.imshow(thresh, cmap='gray'), plt.title(f'Masked Otsu (t={t:.1f})'), plt.axis('off')
    plt.tight_layout()
    plt.show()

    return thresh, t

# Example usage:
# thresh_img, threshold_value = masked_otsu_threshold('gray.png', 'mask.png')
```
